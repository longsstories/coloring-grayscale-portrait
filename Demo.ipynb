```python
from keras.models import load_model
import cv2
import numpy as np

model=load_model('/content/drive/MyDrive/Color_photo/models/last/unet11.h5')
gray_path="/content/1445123.jpg"
gray = cv2.resize(cv2.imread(gray_path), (128,128))
if gray.shape[2]>=3: 
  gray = gray.mean(axis=-1)
gray = gray/255
gray_reshape = gray.reshape((1,128,128,1))

color=model.predict(gray_reshape)[0]
```

```python
from google.colab.patches import cv2_imshow
gray_img = cv2.resize(cv2.imread(gray_path),(256,256))
color_img = cv2.cvtColor(color*255, cv2.COLOR_BGR2RGB)
color_img = cv2.resize(color_img,(256,256))
cv2_imshow(gray_img)
cv2_imshow(color_img)
cv2.waitKey()
```
